```python
content_recs = [
    "The Matrix",
    "Interstellar",
    "Inception"
]

collab_recs = [
    "Toy Story 2",
    "Monsters Inc",
    "Finding Nemo"
]
```

```python
hybrid_recs = list(
set(content_recs + collab_recs)
)

print(hybrid_recs)
```

```python
import pickle

pickle.dump(
    hybrid_recs,
    open("/content/movies.pkl","wb")
)
```
