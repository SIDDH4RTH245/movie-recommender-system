```python
import pandas as pd
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
```

```python
df = ratings.merge(movies, on="movieId")
```

```python
movie_matrix = df.pivot_table(
    index="userId",
    columns="title",
    values="rating"
)
```

```python
toy_story_ratings = movie_matrix["Toy Story (1995)"]
```

```python
similar_movies = movie_matrix.corrwith(
    toy_story_ratings
)
```

```python
corr_df = pd.DataFrame(
    similar_movies,
    columns=["Correlation"]
)

corr_df.dropna(inplace=True)

corr_df.sort_values(
    "Correlation",
    ascending=False
).head(10)
```
