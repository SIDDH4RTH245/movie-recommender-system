```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("movies_cleaned.csv")
```

```python
cv = CountVectorizer(max_features=5000, stop_words="english")

vectors = cv.fit_transform(df["tags"]).toarray()
```

```python
similarity = cosine_similarity(vectors)
```

```python
matching_movies = df[df["title"]=="Inception"]

if not matching_movies.empty:
    movie_index = matching_movies.index[0]
    distances = similarity[movie_index]

    recommended_movies_with_distances = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:4]
    recommended_movie_indices = [i[0] for i in recommended_movies_with_distances]
    recommended_titles = df.loc[recommended_movie_indices, "title"].tolist()

    print(f"Recommended movies for 'Inception': {recommended_titles}")
else:
    print("Error: Movie 'Inception' not found in the dataset. Please check the exact title.")
    # You might want to list some similar titles or provide a way to search if this is a common issue.
    # For example, if 'Inception (2010)' is in the dataset, you could suggest it:
    # if "Inception (2010)" in df["title"].values:
    #     print("Did you mean 'Inception (2010)'?")
```
